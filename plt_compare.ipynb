```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import params
```

```python
checkpoint_period = params.log_period
path = 'Dueling_Double_DQN/results_different_seeds/'

seed1=np.loadtxt(path + 'seed1/seed1_rewards.txt', skiprows=0)
seed2=np.loadtxt(path + 'seed2/rewards2.txt', skiprows=0)
seed3=np.loadtxt(path + 'seed3/rewards2.txt', skiprows=0)
print (seed1.shape, seed2.shape, seed3.shape)
```

```python
avg = np.mean((seed1,seed2,seed3),axis=0)
std = np.std((seed1,seed2,seed3), axis=0)
print (avg.shape, std.shape)
```

```python
#Thing this is moving average with window of 10
from scipy.ndimage.filters import uniform_filter1d

checkpoint_period = 10

episodes = avg[:,0]*checkpoint_period
mean = uniform_filter1d(avg[:,1], size=10)
mean_std = uniform_filter1d(std[:,1], size=10)
max = uniform_filter1d(avg[:,2], size=10)
max_std = uniform_filter1d(std[:,2], size=10)

matplotlib.rc('font',family='Times New Roman')
matplotlib.rcParams['font.size'] = 20
fig, ax = plt.subplots(1,1,figsize=(10,8))

ax.plot(episodes,mean, label="Average")
ax.plot(episodes,max, 'C3-', label="Max")

ax.fill_between(episodes,mean-mean_std, mean+mean_std, color='blue', alpha=0.1)
ax.fill_between(episodes,max-max_std, max+max_std, color='red', alpha=0.1)

ax.set_ylabel("Reward")
ax.set_xlabel("Episodes")

yticks = np.arange(0,450,50)
plt.yticks(yticks)
plt.tight_layout()
plt.legend(loc='upper left')
plt.title("Dueling Double DQN")
ax.grid()
```
